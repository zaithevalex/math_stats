```python
def round_list_of_tuples(list_of_tuples, decimals=2):
    """Rounds the numbers in a list of tuples"""
    new_list = []

    for tpl in list_of_tuples:
        new_tuple = (round(tpl[0], decimals), round(tpl[1], decimals))
        new_list.append(new_tuple)

    return new_list
```
