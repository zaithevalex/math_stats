```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv('free_throws.csv')
print(df.shape)
df.head( )
```

```python
dff = df[['game', 'player']].drop_duplicates()
dct = dff.groupby('player')['game'].agg(lambda w: list(set(w))).to_dict()
```

```python
from collections import Counter
command = dict()

for k,v in dct.items():
    items = dct[k]
    items  = [jtem for i in items for jtem in i.split(' - ')]
    com = [item[0] for item in Counter(items).most_common()]
    command[k] = com
```

```python
x = df.player.values
y = df.game.apply(lambda w: w.split(' - ')).values

cnt = 0
z = [ ]
for player, com in zip(x, y):
    cc = command[player]
    is_it = True
    for i in cc:
        if i in com:
            z.append(i)
            is_it = False
            break
    if is_it:
        z.append(None)
        cnt += 1  
```

```python
df['command'] = z
df['is_home'] = df.game.apply(lambda w: w.split(' - ')[0]) == df.command
```
